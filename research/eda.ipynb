# EDA and Feature Engineering

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

# Explorary Data Analysis

### Dataset Overview

```python
# please change this line to your own data directory
folder_path = "/Users/kang/Downloads/optiver-additional-data"

feature_path = os.path.join(folder_path, "order_book_feature.parquet")
target_path = os.path.join(folder_path, "order_book_target.parquet")
trade_path = os.path.join(folder_path, "trades.parquet")
time_id_path = os.path.join(folder_path, "time_id_reference.csv")

trade_df = pd.read_parquet(trade_path, engine='pyarrow', columns=None)
time_id_df = pd.read_csv(time_id_path, sep="\t")
```

### Feature Dataset: `order_book_feature.parquet`

```python
feature_df = pd.read_parquet(feature_path, engine='pyarrow', columns=None)

print("There are", feature_df["stock_id"].nunique(), "differennt stocks in the feature dataset:", feature_df["stock_id"].unique())
print("Across", feature_df["time_id"].nunique(), "time_ids:", feature_df["time_id"].unique())
print("Time Range:", feature_df["seconds_in_bucket"].min(), "-", feature_df["seconds_in_bucket"].max())

sample = feature_df[(feature_df["stock_id"] == 8382) & (feature_df["time_id"] == 100)].sort_values("seconds_in_bucket")

plt.plot(sample["seconds_in_bucket"], sample["bid_price1"], label="Bid Price 1")
plt.plot(sample["seconds_in_bucket"], sample["ask_price1"], label="Ask Price 1")
plt.title("Bid-Ask Price Over Time (Stock 8382)")
plt.xlabel("Seconds in Bucket")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```

### Target Dataset: `order_book_target.parquet`

```python
target_df = pd.read_parquet(target_path, engine='pyarrow', columns=None)

print("There are", target_df["stock_id"].nunique(), "differennt stocks in the feature dataset:", target_df["stock_id"].unique())
print("Across", target_df["time_id"].nunique(), "time_ids:", target_df["time_id"].unique())
print("Time Range:", target_df["seconds_in_bucket"].min(), "-", target_df["seconds_in_bucket"].max())

sample = target_df[(target_df["stock_id"] == 8382) & (target_df["time_id"] == 100)].sort_values("seconds_in_bucket")

plt.plot(sample["seconds_in_bucket"], sample["bid_price1"], label="Bid Price 1")
plt.plot(sample["seconds_in_bucket"], sample["ask_price1"], label="Ask Price 1")
plt.title("Bid-Ask Price Over Time (Stock 8382)")
plt.xlabel("Seconds in Bucket")
plt.ylabel("Price")
plt.legend()
plt.grid(True)
plt.show()
```
